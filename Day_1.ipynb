```python
# Small Python program showing variables and data types

# Integer
age = 24
print("Age:", age, "Type:", type(age))

# Float
height = 5.9
print("Height:", height, "Type:", type(height))

# String
name = "Venkat"
print("Name:", name, "Type:", type(name))

# Boolean
is_student = False
print("Is Student:", is_student, "Type:", type(is_student))

# List
marks = [85, 90, 78]
print("Marks:", marks, "Type:", type(marks))

# Tuple
coordinates = (10, 20)
print("Coordinates:", coordinates, "Type:", type(coordinates))

# Dictionary
person = {"name": "Venkat", "age": 24, "student": False}
print("Person:", person, "Type:", type(person))
```
